```python
## In the dataset. the column Sub-sample wise multiplier is used to estimate the population size for each level of survey unit.
## Documentataion has to be refrred to understend the methodlogy in depth.

## As per methodology, in this code block, if NS Count for sub-sample (given as NSS in the documentation) equals to NS count for
## substratum(given as NSO in the documentation) then to estimate the population in the survey unit multiplier column has to be devide by 400
## other wise it has to be devide by 800.

## using this information  this code block first create a function which apply the above logic. the function has bene named as calculate_final_weight.
## this function has been applied on the dataset to create a new column termed here as 'final_weight'
def calculate_final_weight(row):
    if row['Ns_count_for_sector_x_Statum_x_substratum_x_sub-sample'] == row['Ns_count_for_sector_x_stratum_x_substratum']:
        return row['Sub-sample_wise_multiplier'] / 400
    else:
        return row['Sub-sample_wise_multiplier'] / 800

# Apply the function to each row to create the new column
df_perv1_above15['final_weight'] = df_perv1_above15.apply(calculate_final_weight, axis=1)
```

```python
# calculating the Labour force for Usual Status.
# To understand what is usual status, please refer to the documentaion (survey design and methodology)

# the codes stored in variable 'status_codes_in_LF_US' belongs to the persons who are in workfoce
# (different codes signifies diiferent nature of engagement but in the labour force)

# A person is considered in workforce if the person is enagaged in any form which belong to the following code or he is engaged in any susidiary work.
# The code below apply the same logic. It filters out all the population as per the above logic and then sums of all the estimated population (from
# final_weight column) belongin to these catroies. This gives us total population in the absolute number.
status_codes_in_LF_US = ['11','12', '21', '31', '41', '42', '51', '61', '71', '72', '81', '82']
Total_Labour_force_US = df_perv1_above15.loc[
    (df_perv1_above15['Status_code_princ'].isin(status_codes_in_LF_US)) |
    (df_perv1_above15['Whether_engaged_in_any_work_in_sussidiary_capacity'] == 1),
    'final_weight'].sum()

# The below code devides the total population in the labour force from the total population and multiply it by 100 to get the percentage LFPR in usual status
LFPR_US = Total_Labour_force_US*100/ df_perv1_above15['final_weight'].sum()
print(LFPR_US)
```

```python
## SImilar process as above have been applied to calculate the LFPR in CWS category. to understand CWS (Current weekly status, please refer documentation)
## A simple difference here is that engagement in subsidiary work does not apply in case of CWS that why
## has not been used to calculate the total no. of persons in LFPR in CWS.
status_codes_in_LF_CWS = ['11','12', '21', '31', '41', '42', '51', '61', '62','71', '72', '81', '82']
Total_Labour_force_CWS = df_perv1_above15.loc[(df_perv1_above15['Current_weekly_status_CWS'].isin(status_codes_in_LF_CWS)),'final_weight'].sum()

LFPR_CWS = Total_Labour_force_CWS*100/ df_perv1_above15['final_weight'].sum()
print(LFPR_CWS)
```
